```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://ppfocus.com/0/asd29276a.html'
```

# 爬找算命網站


```python
resp=requests.get(url)
resp
```

```python
resp.text
```

```python
soup=BeautifulSoup(resp.text,'lxml')
soup
```

```python
soup
```

```python
ps=soup.find(id="js_content").find_all('p')
ps
```

```python
len(ps)
```

```python
ps[31].text.strip().replace('、',' ').split()
```

```python
datas=[]
for p in ps[31:112]:
    data=[]
    data.append(p.text.strip().replace('、',' ').split())
    datas.extend(data)
print(datas)
```
